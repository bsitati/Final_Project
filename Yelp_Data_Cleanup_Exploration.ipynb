## Extract CSVs into DataFrames

```python
businessData = pd.read_json('https://yelp-business-bucket.s3.amazonaws.com/yelp_academic_dataset_business.json', lines=True)
businessData.head(50)
```

```python
businessData = businessData.dropna(axis=0)
businessData = businessData.reset_index(drop=True)
businessData.shape  # Check number of rows and columns
```

```python
# Check if business contains keyword "Restaurants"
businessData["is_restaurant"] = businessData["categories"].str.contains("Restaurants") 
businessData["is_restaurant"] =businessData["is_restaurant"].replace({True: 1, False: 0})# Convert True/ False to 0/ 1

businessData[businessData.is_restaurant == 1] #keep only businesses that are restaurants
```

```python
restaurantData = businessData
restaurantData = restaurantData.reset_index(drop=True)
```

```python
attributes = ['Alcohol',
              'Ambience',
              'BikeParking',
              'BusinessParking',
              'GoodForKids',
              'HasTV',
              'OutdoorSeating',
              'RestaurantsAttire',
              'RestaurantsPriceRange2',
              'BusinessAcceptsCreditCards',
              'RestaurantsReservations',
              'RestaurantsGoodForGroups',
              'RestaurantsDelivery',
              'WiFi']

counter = 0
for i in range(0,len(restaurantData)):
    
    ats = [i.split(':', 1)[0] for i in restaurantData['attributes'][i]]
    contain = [i for i in ats if i in attributes]
    
    if len(contain) == len (attributes):
            counter += 1

print("Number of restaurants containing all attributes:  ", counter)
print("Total Number of restaurants:  ", len(restaurantData))
```

```python
# Get Parking
df_attribute = pd.DataFrame([i if i != None else {'BusinessParking':np.NAN} for i in restaurantData3['attributes']])
restaurantData3['parking'] = df_attribute['BusinessParking']

parking = restaurantData3['parking']
```

```python
# Convert parking string with dictionary into dictionary
def json_to_df(obs):
    obs = obs.replace("\'", "\"")
    obs_dict = eval(obs)

    return(pd.DataFrame([obs_dict]))

# Create empty dataframe with columns
parking_df = pd.DataFrame(columns=['garage', 'street', 'validated', 'lot', 'valet'])

# Iterate over pandas series
#...consider using apply
for item in parking.iteritems():
    temp_df = json_to_df(item[1])
    parking_df = parking_df.append(temp_df)
    
parking_df
```

```python
cleaned_parking_df = parking_df.drop(0, axis = 1)
cleaned_parking_df.reset_index(drop = True, inplace = True)
cleaned_parking_df
```

```python
parking_columns=['parking_garage', 'parking_lot', 'parking_street', 'parking_valet', 'parking_validated']
restaurantData3[parking_columns] = cleaned_parking_df
#restaurantData3.head(20)
```

## Calculate Restaurant Density

```python
df2=restaurantData3

for index1, row1 in df2.iterrows():
    num_restaurants = 0
    min_lat = row1['latitude']-.07/10
    max_lat = row1['latitude']+.07/10
    min_long = row1['longitude']-.09/10
    max_long = row1['longitude']+.09/10
    df_temp = df2[(df2['latitude']>min_lat) & 
                  (df2['latitude']<max_lat) & 
                  (df2['longitude']>min_long) & 
                  (df2['longitude']<max_long)]
    num_restaurants = len(df_temp)
    df2.loc[index1, 'num_restaurants_1mile'] = num_restaurants - 1 # (subtract 1 to exclude self)"
    
    
df2.head()
```

```python
den_category = ['density_<=8', 'density_<=18', 'density_<=38', 'density_<=274'] # Pass bin name
# Create density bins
restaurantData3['den_category']=pd.cut(x= restaurantData3['num_restaurants_1mile'], bins=[0, 8, 18, 38, 274],labels=den_category)

# *********Dummified binned density**************
review_df = pd.get_dummies(restaurantData3["den_category"])
review_df.head()

restaurantData3[den_category] = review_df

#restaurantData3.head(20)
```
